```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Extracting Data
EPOCHS = 12
filenames = ["CIFAR10 Benchmark, T=1, {}","cifar10 T=half, {}","cifar10 T=third, {}", "cifar10 T=quarter, {}","cifar10 T=fifth, {}","cifar10 T=sixth, {}"]
Data_test = []
Data_train = []
Tvals = []

for j in range(len(filenames)):
    Data_test.append(np.loadtxt(filenames[j].format('test accuracy.txt')))
    Data_train.append(np.loadtxt(filenames[j].format('train accuracy.txt')))
    Tvals.append([1/(j+1)**k for k in range(EPOCHS)])
    
Data_evaluation=[]
for j in range(6):
    Data_evaluation.append(Data_test[j][12])
```

```python
figlabels = ["cifar10 Benchmark, T=1","cifar10 T=1/2^(EPOCH)","cifar10 T=1/3^(EPOCH)","cifar10 T=1/4^(EPOCH)","cifar10 T=1/5^(EPOCH)","cifar10 T=1/6^(EPOCH)"]

for j in range(len(filenames)):
    plt.plot(range(EPOCHS),Data_test[j][0:EPOCHS],label = figlabels[j])
   
plt.title('Test Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('test accuracy')
plt.legend()
```

```python
for j in range(len(filenames)):
    plt.plot(range(EPOCHS),Data_train[j],label = figlabels[j])
   
plt.xlabel('Epoch')
plt.ylabel('train accuracy')
plt.title('Training Accuracy per Epoch')
plt.legend()
```

```python
plt.plot(range(len(filenames)), Data_evaluation)
plt.title('Evaluation after training')
plt.xlabel('T=1/2**x')
plt.ylabel('Test Accuracy at "Room Temperature"')
```
